```python
#importing pandas,numply,matplotlib.pyplot,seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading 'iris' data into a pandas dataframe from a excel file
data=pd.read_excel('iris.xls')
```

```python
#Display the columns in the dataset
data.columns
```

```python
#displaying data types of each column
data.dtypes
```

```python
# mean of "SL" column of the dataset and rounding values to 2 decimal places.
round(data["SL"].mean(),2)
```

```python
# mean of "SW" column of the dataset and rounding values to 2 decimal places.
round(data["SW"].mean(),2)
```

```python
# mean of "PL" column of the dataset and rounding values to 2 decimal places.
round(data["PL"].mean(),2)
```

```python
# mean of "PW" column of the dataset and rounding values to 2 decimal places.
round(data["PW"].mean(),2)
```

```python
#cant take mean of "Classification" ,because its datatype is 'object'
```

```python
# Check for the null values present in the dataset.
data.info()
```

```python
#boxplot grouped by classification
data.boxplot(by="Classification",figsize=(15, 10))
```

```python
# Paired plot using seaborn
sns.pairplot(data,hue="Classification")
```

```python
#heat map
sns.heatmap(data.corr(),annot=True)
```

```python
#Classification count using count plot
sns.countplot(data['Classification'])
```

```python
#violin plot of SW in each classification
sns.violinplot(x='Classification',y='SW',data=data)
```

```python
#scatter plot of SW vs SL in each classification
sns.scatterplot(x='SW', y='SL', hue='Classification', data=data)
```

```python
#scatter plot of PW vs PL in each classification
sns.scatterplot(x='PW', y='PL', hue='Classification', data=data)
```
